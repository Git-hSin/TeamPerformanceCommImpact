```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
mlb_data_file = "Resources/mlb.csv"
mlb_attendance_file = "Resources/mlb_attendance.csv"
mlb_locations_file = "Resources/mlb_loc.csv"
nfl_data_file = "Resources/nfl.csv"
nfl_attendance_file = "Resources/nfl_attendance.csv"
nfl_locations_file = "Resources/nfl_loc.csv"
nba_data_file = "Resources/nba.csv"
nba_attendance_file = "Resources/nba_attendance.csv"
nba_locations_file = "Resources/nba_loc.csv"
```

```python
mlb_df = pd.read_csv(mlb_data_file, low_memory = False)
mlb_attendance_df = pd.read_csv(mlb_attendance_file, low_memory = False)
mlb_locations_df = pd.read_csv(mlb_locations_file, low_memory = False)
nfl_df = pd.read_csv(nfl_data_file, low_memory = False)
nfl_attendance_df = pd.read_csv(nfl_attendance_file, low_memory = False)
nfl_locations_df = pd.read_csv(nfl_locations_file, low_memory = False)
nba_df = pd.read_csv(nba_data_file, low_memory = False)
nba_attendance_df = pd.read_csv(nba_attendance_file, low_memory = False)
nba_locations_df = pd.read_csv(nba_locations_file, low_memory = False)
```

```python
def nfl_team(x):
   nfl_data=nfl_df[nfl_df["Tm"] == x].reset_index(drop=True)
   return(nfl_data)
def nba_team(x):
   nba_data=nba_df[nba_df["Team"]==x].reset_index(drop=True)
   return(nba_data)
def mlb_team(x):
   mlb_data=mlb_df[mlb_df['team']==x].reset_index(drop=True)
   return(mlb_data)
```

# Top/Middle/Bottom 3 performing teams over 20 years

## Top 3 MLB Teams

```python
mlb_team_list = []
mlb_team_total_win = []
mlb_team_total_loss = []
for team in mlb_df["team"].unique():
    mlb_team_list.append(team)
    team_data = mlb_team(team)
    mlb_team_total_win.append(team_data["win"].sum())
    mlb_team_total_loss.append(team_data["loss"].sum())
mlb_team_totals_df = pd.DataFrame({
    "team" : mlb_team_list,
    "total_wins": mlb_team_total_win,
    "team_losses": mlb_team_total_loss
})
mlb_team_totals_df.sort_values(by="total_wins",ascending=False).set_index("team")
```

### Top 3 Teams: NYY, BOS, STL

## Median 3 MLB Teams

```python
np.median(mlb_team_total_win)
```
